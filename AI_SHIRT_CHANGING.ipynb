```python
import cv2 as cv
import cvzone
from cvzone.PoseModule import PoseDetector
import os
```

```python
cap=cv.VideoCapture(0)
detector=PoseDetector()
listshirts=os.listdir('shirts/')
fixedratio=278/251 # width of shirt/widthofpoint11 and 12
shirtratioheightwidth=581/440
while True:
    ret,frame=cap.read()
    pose=detector.findPose(frame)
#     pose=cv.flip(pose,1)
    lmList,bboxInfo=detector.findPosition(pose,bboxWithHands=False,draw=False)
    
    if bboxInfo:
        center=bboxInfo['center']
        lm11=lmList[11][1:3]
        lm12=lmList[12][1:3]
        
        shirtimg=cv.imread(os.path.join('shirts/',listshirts[0]),cv.IMREAD_UNCHANGED)
        widthofshirt=int ((lm11[0]-lm12[0])*fixedratio)
        print(widthofshirt)
        if widthofshirt > 0 and int(widthofshirt * shirtratioheightwidth) >0:
            shirtimg = cv.resize(shirtimg, (widthofshirt, int(widthofshirt * shirtratioheightwidth)))
    #         shirtimg = cv.resize(shirtimg, (lm11[0],lm12[0]))
            current_scale=widthofshirt/190
            offset=int(current_scale*44),int(current_scale*48)
            try:
                frame = cvzone.overlayPNG(frame, shirtimg, (lm12[0]-offset[0],lm12[1]-offset[1]))
            except:
                pass
        

    cv.imshow('frame',frame)
    key=cv.waitKey(2)
    if key==ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```
