```python
import cv2 as cv
import numpy as np
```

```python
def setValue(x):
    print('')
cap=cv.VideoCapture(0)

window=cv.namedWindow('window')
cv.createTrackbar('upper hue','window',110,180,setValue)
cv.createTrackbar('upper saturation','window',255,255,setValue)
cv.createTrackbar('upper value','window',255,255,setValue)
cv.createTrackbar('lower hue','window',60,180,setValue)
cv.createTrackbar('lower saturation','window',55,255,setValue)
cv.createTrackbar('lower value','window',54,255,setValue)

while True:
    cv.waitKey(1000)
    ret,init_frame=cap.read()#intial frame
    if(ret):
        break
while True:
    ret,frame=cap.read()
    hsv=cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    
    u_hue=cv.getTrackbarPos('upper hue','window')
    u_saturation=cv.getTrackbarPos('upper saturation','window') 
    u_value=cv.getTrackbarPos('upper value','window')      
    l_hue=cv.getTrackbarPos('lower hue','window')         
    l_saturation=cv.getTrackbarPos('lower saturation','window')         
    l_value=cv.getTrackbarPos('lower value','window')   
    
    kernel=np.ones((3,3),np.uint8)

    upper_hsv=np.array([u_hue,u_saturation,u_value])
    lower_hsv=np.array([l_hue,l_saturation,l_value])
    
    mask=cv.inRange(hsv,lower_hsv,upper_hsv)
    mask=cv.medianBlur(mask,3)
    mask_inv=255-mask
    mask=cv.dilate(mask,kernel,5)
    
    b=frame[:,:,0]
    g=frame[:,:,1]    
    r=frame[:,:,2]
    b=cv.bitwise_and(mask_inv,b)
    g=cv.bitwise_and(mask_inv,g)
    r=cv.bitwise_and(mask_inv,r)
    frame_inv=cv.merge((b,g,r))
    
    b=init_frame[:,:,0]
    g=init_frame[:,:,1]    
    r=init_frame[:,:,2]
    b=cv.bitwise_and(b,mask)
    g=cv.bitwise_and(g,mask)
    r=cv.bitwise_and(r,mask)
    blanketarea=cv.merge((b,g,r))
    
    final=cv.bitwise_or(frame_inv,blanketarea)
    cv.imshow('harryPotter',final)
#     cv.imshow('mask',mask)
#     cv.imshow('mask_inv',mask_inv)
#     cv.imshow('initframe',init_frame)
#     cv.imshow('originalframe',frame)
    
    
    
    if (cv.waitKey(3)==ord('q')):
        break
cv.destroyAllWindows()
cap.release()
    
    
```
